```
import random
```

```
words_with_hints = {
    "python": "A popular programming language",
    "machine": "Related to AI and learning",
    "hangman": "Classic word guessing game",
    "internet": "Connects computers worldwide",
    "algorithm": "Step-by-step problem-solving method"
}

word = random.choice(list(words_with_hints.keys()))
hint = words_with_hints[word]
```

```
guessed_letters = []
attempts = 6
display_word = ["_"] * len(word)
```

```
stages = [
    """
      ------
      |    |
      |    O
      |   \\|/
      |    |
      |   / \\
      -
    """,
    """
      ------
      |    |
      |    O
      |   \\|/
      |    |
      |   /
      -
    """,
    """
      ------
      |    |
      |    O
      |   \\|/
      |    |
      |
      -
    """,
    """
      ------
      |    |
      |    O
      |   \\|
      |    |
      |
      -
    """,
    """
      ------
      |    |
      |    O
      |    |
      |    |
      |
      -
    """,
    """
      ------
      |    |
      |
      |
      |
      |
      -
    """,
    """
      ------
      |
      |
      |
      |
      |
      -
    """
]
```

```
print("Welcome to Hangman Game!")
print("Hint:", hint)
while attempts > 0 and "_" in display_word:
    print(stages[attempts])
    print("Word:", " ".join(display_word))
    print("Guessed letters:", guessed_letters)
    guess = input("Guess a letter: ").lower()
    if guess in guessed_letters:
        print("âš  Already guessed!")
        continue
    guessed_letters.append(guess)
    if guess in word:
        print("Correct guess!")
        for i in range(len(word)):
            if word[i] == guess:
                display_word[i] = guess
    else:
        print("Wrong guess!")
        attempts -= 1
```

```
if "_" not in display_word:
    print("\nCongratulations! You guessed the word:", word)
else:
    print(stages[0])
    print("\nGame Over! The word was:", word)
```
